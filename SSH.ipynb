# SSH
* Secure Sheel
* Communication Protocol (Like http, https, ftp, etc)
* Do just about anything on the remote computer
* Traffic is encrypted
* Used mostly in the terminal/command line

## Client/Server Comunication
* SSH is the client
* SSHD is the server (Open SSH Daemon)
* The server must have sshd installed and running or you will not be able to connect using ssh

## Authentication Methods
> ssh brad@192.168.1.29
* Password
* Public/Private Key Pair (saver, no brutforce possible)
* Host Based

## Generating Keys
> ssh-keygen
* ~/.ssh/id_rsa(Private Key)
* ~/.ssh/id_rsa.pub(Public Key)
* Public key goes into server "authorized_keys" file

## Add keys to brads .ssh on the server

```python
# log back in as root
ssh root@doserver
cd /home/brad
mkdir .ssh
cd .ssh
touch authorized_keys
sudo nano authorized_keys (past in the id_rsa_do.pub key)
# exit and log in as brad
```

## Disable root password login

```python
sudo nano /etc/ssh/sshd_config
# set the following
PermitRootLogin no
PasswordAuthentication no
# reload sshd service
sudo systemctl reload sshd
# changw owner of/home/brad/* to brad
sudo chown -R brad:brad /home/brad
# may need to set permmission
chmod 700 /home/brad/.ssh
```
